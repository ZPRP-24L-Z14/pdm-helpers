```python
import json
 
# Opening JSON file
f = open('top-pypi.json')
 
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
with open("modules.txt", "w") as file:
    for d in data["rows"]:
        file.write('\"' + d["project"] + '\"'+ "\n")
```

```python
a = []
with open("modules.txt", "r") as file:
    for row in file.readlines():
        a.append(row)
```
